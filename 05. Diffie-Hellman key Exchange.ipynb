```python
# Given public keys P and G
P = 31
G = 3

# Randomly selected private keys a and b
a = 8
b = 5

# Calculate shared keys exchanged between Alice and Bob
alice_y = pow(G, a, P)
bob_x = pow(G, b, P)
print(alice_y, bob_x)

# Calculate secret keys for Alice and Bob
Ka = pow(bob_x, a, P)
Kb = pow(alice_y, b, P)
print(Ka, Kb)

# Verify if Ka is equal to Kb
if Ka == Kb:
    print("Ka is equal to Kb. Key exchange successful!")
else:
    print("Ka is not equal to Kb. Key exchange failed.")
```
