```python
from sympy.ntheory.modular import crt

def calculate_M(c1, c2, c3, n1, n2, n3, e):
    # CRT to solve the congruences
    N = [n1, n2, n3]
    c = [c1, c2, c3]
    x = crt(N, c)[0]

    # Calculate the cube root of x to get M
    M = round(pow(x, 1 / e))

    return M

# Given ciphertext and modulus
c1 = 1632893752
c2 = 1226329111
c3 = 425882503
n1 = 2015805937
n2 = 3148630007
n3 = 1883043919

# Given encryption exponent
e = 3

M = calculate_M(c1, c2, c3, n1, n2, n3, e)
print("Plaintext M:", M)
```

```python
def gcd_extended(a, b):
    if a == 0:
        return (b, 0, 1)
    gcd, x1, y1 = gcd_extended(b % a, a)
    x = y1 - (b // a) * x1
    y = x1
    return (gcd, x, y)

def modular_inverse(a, m):
    gcd, x, y = gcd_extended(a, m)
    if gcd != 1:
        raise ValueError("The modular inverse does not exist.")
    return x % m

def calculate_M(c1, c2, c3, n1, n2, n3, e):
    # Calculate n1 * n2 * n3
    N = n1 * n2 * n3

    # Calculate the sum of the remainders using CRT
    sum_remainders = (c1 * n2 * n3 * modular_inverse(n2 * n3, n1) +
                      c2 * n1 * n3 * modular_inverse(n1 * n3, n2) +
                      c3 * n1 * n2 * modular_inverse(n1 * n2, n3)) % N

    # Calculate the cube root of the sum of remainders to get M
    M = round(sum_remainders ** (1 / e))

    return M

# Given ciphertext and modulus
c1 = 1632893752
c2 = 1226329111
c3 = 425882503
n1 = 2015805937
n2 = 3148630007
n3 = 1883043919

# Given encryption exponent
e = 3

M = calculate_M(c1, c2, c3, n1, n2, n3, e)
print("M:", M)
```
